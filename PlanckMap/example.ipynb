# The Planck CMB (perceptually uniform)

```python
import matplotlib
import matplotlib.pyplot as plt

import healpy as hp
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# https://irsa.ipac.caltech.edu/data/Planck/release_1/all-sky-maps/index.html
data = hp.read_map("COM_CompMap_CMB-smica_2048_R1.20.fits", ("INP_CMB",))
```

```python
cmap_name = "RdYlBu_r"
xsize, ysize, dpi = 15, 10, 100
```

```python
cmap = matplotlib.cm.get_cmap(cmap_name).copy()
cmap.set_bad("gray") # missing pixels
cmap.set_under("white") # background
```

```python
fig = plt.figure(figsize=(xsize, ysize))

hp.mollview(data,
    xsize=xsize*dpi,
    min=-400,
    max=400,
    title="",
    cbar=True,
    cmap=cmap,
    unit=r"$\delta$T ($\mu$K)"
)

plt.savefig(f"planck_cmb_{cmap_name}.png", dpi=dpi, bbox_inches="tight")
plt.show()
```
